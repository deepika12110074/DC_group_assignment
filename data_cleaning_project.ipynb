```python
# import required modules
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import json

def getPageURLs(df, columns, url, visited_urls, p):
    # get URL
    j = 1
    print("Loading Wiki NASDAQ page "+str(j)+"....")
    j += 1
    page = requests.get("https://en.wikipedia.org"+url)
    
    visited_urls.append("https://en.wikipedia.org"+url)
    # display status code
    print(page.status_code)
    
    # scrape webpage
    soup = BeautifulSoup(page.content, 'html.parser')
    
    # create object
    object = soup.find(id="mw-pages")
    
    # find tags
    items = object.find_all(class_="mw-category")
    result = items[0]

    i = 0
    k = 1
    company_list = result.find_all('ul')
    
    for cmp in company_list:
        a_tags = result.find_all('ul')[i].find_all('a')
        for at in a_tags:
            TextToWrite = []
            link = at['href']
            compname = at['title']
            
            print("Loading data from "+compname+"....")
            response = requests.get("https://en.wikipedia.org"+link)
            
            visited_urls.append("https://en.wikipedia.org"+link)
            
            soup = BeautifulSoup(response.text,'html.parser')
            findtable = soup.find('table',{'class':'infobox vcard'})
            if (findtable):
                table = findtable.tbody
                rows = table.find_all('tr')
                values = { "data":[]}
                #compdata = {'S.No.': i, 'Company Name':compname, 'Type':'', 'Traded as':'', 'ISIN':'', 'Industry':'', 'Founded':'', 'Founders':'', 'Headquarters':'', 'Number of locations':'', 'Area served':'', 'Products':'', 'Services':'', 'Revenue':'', 'Operating income':'', 'Net income':'', 'Total assets':'', 'Total equity':'', 'Key people':'', 'Number of employees':'', 'Subsidiaries':'', 'Divisions':'', 'Website':''}
                compdata = {}
                compdata['Company Name'] = compname
                for tr in range(0, len(rows)):
                    th = rows[tr].find('th')
                    if th != None:
                        key = th.text.replace('\n','')
                        if(key == 'Founder' or key == 'Founder(s)'):
                            key = "Founders"
                        elif(key == 'Employees'):
                            key = "Number of employees"
                        elif(key == 'Launched'):
                            key = "Founded"
                        elif(key == 'URL'):
                            key = "Website"
                        elif(key == 'Type of business' or key == 'Type of site'):
                            key = "Type"
                        else:
                            key = key
                        if key in columns:
                            td = rows[tr].find('td').text.replace('\n',''.replace('\xa0',''))
                            compdata[key] = td
                        else:
                            pass;
            else:
                compdata = {}
                compdata['Company Name'] = compname


            values["data"].append(compdata)

            # write the data
            df = df.append(values["data"], ignore_index = True)
            #df = df.append(values["data"], ignore_index=False)
            time.sleep(2)
        i = i+1
    next_page = object.find_all("a")[p]['href']
    exists = next_page.find("pagefrom")
    if(exists != -1):
        getPageURLs(df, columns, next_page, visited_urls, 2)
    else:     
        # Saving visited URLs
        with open('crawled_URLs.json', 'w+') as f:
            json.dump(visited_urls, f)
        df.to_csv("nasdaq_companies.csv")
        return visited_urls

visited_urls = []
main_url = "/w/index.php?title=Category:Companies_listed_on_the_Nasdaq"
columns = ['Company Name', 'Type', 'Traded as', 'ISIN', 'Industry', 'Founded', 'Founders', 'Headquarters', 'Number of locations', 'Area served', 'Products', 'Services', 'Revenue', 'Operating income', 'Net income', 'Total assets', 'Total equity', 'Key people', 'Number of employees', 'Subsidiaries', 'Divisions', 'Website']
df = pd.DataFrame(columns=columns)
#df.set_index(['S.No.'], inplace=True)
df = getPageURLs(df, columns, main_url, visited_urls, 1)
```
